```python
from __future__ import division, print_function
%matplotlib inline
from sklearn import datasets,linear_model
from scipy import optimize
import numpy as np
import matplotlib.pyplot as plt
from logistic_regression import LogisticRegression
```

# Demonstrating LogisticRegression. Comparison LogisticRegression with sklearn.linear_model.LogisticRegression

###### Loading dataset and preparing test and training sets

```python
# import some data to play with
iris = datasets.load_iris()
X = iris.data[iris.target<2, :2]  # we only take the first two features.
Y = iris.target[iris.target<2]    # we only take examples of 1 and 0  

h = .02  # step size in the mesh
```

###### Demonstrating LogisticRegression

```python
# Create logistic regression object with custom model
log_regr=LogisticRegression(0.3,0.35)

#fitting data
J,theta=log_regr.gradient_descent(X,Y)
```

```python
# Plot the decision boundary. For that, we will assign a color to each
# point in the mesh [x_min, x_max]x[y_min, y_max].
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

set_for_prediction=np.vstack((xx.ravel(), yy.ravel())).T

Z = log_regr.predict(set_for_prediction)

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.figure(1, figsize=(4, 3))
plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Paired)

# Plot also the training points
plt.scatter(X[:, 0], X[:, 1], c=Y, edgecolors='k', cmap=plt.cm.Paired)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.xticks(())
plt.yticks(())

plt.show()
```

###### Comparison with sklearn.linear_model.LogisticRegression

```python
logreg = linear_model.LogisticRegression(C=1e5)

# we create an instance of Neighbours Classifier and fit the data.
logreg.fit(X, Y)

# Plot the decision boundary. For that, we will assign a color to each
# point in the mesh [x_min, x_max]x[y_min, y_max].
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = logreg.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.figure(1, figsize=(4, 3))
plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Paired)

# Plot also the training points
plt.scatter(X[:, 0], X[:, 1], c=Y, edgecolors='k', cmap=plt.cm.Paired)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.xticks(())
plt.yticks(())

plt.show()
```
