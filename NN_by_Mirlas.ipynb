```python
from __future__ import division, print_function
%matplotlib inline
from sklearn import datasets,neural_network
from scipy import optimize
import numpy as np
import matplotlib.pyplot as plt
from NN import NeuralNetwork
```

# Demonstrating NeuralNetwork. Comparison NeuralNetwork with sklearn.neural_network.MLPClassifier

###### Loading dataset and preparing test and training sets

```python
# Generate a dataset and plot it
np.random.seed(0)
X, Y = datasets.make_moons(200, noise=0.20)
plt.scatter(X[:,0], X[:,1], s=40, c=Y, cmap=plt.cm.Spectral)
```

```python
y = np.ones((Y.shape[0],len(np.unique(Y))))
for i in range(len(Y)):
    y[i] = (np.array([0,1])==Y[i]).astype(int)
    
h = .02
```

###### Demonstrating NeuralNetwork

```python
# Create logistic regression object with custom model
NN_model=NeuralNetwork((2,10,10,2),learn_rate=20,reg_coef=0.01,eps=0.00001,max_iter=50000)
    
#fitting data
W,J=NN_model.train(X,y)
```

```python
# Plot the decision boundary. For that, we will assign a color to each
# point in the mesh [x_min, x_max]x[y_min, y_max].
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

set_for_prediction=np.vstack((xx.ravel(), yy.ravel())).T

Z = NN_model.predict(set_for_prediction)

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.figure(1, figsize=(4, 3))
plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Paired)

# Plot also the training points
plt.scatter(X[:, 0], X[:, 1], c=Y, edgecolors='k', cmap=plt.cm.Spectral)

plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.xticks(())
plt.yticks(())

plt.show()
```
